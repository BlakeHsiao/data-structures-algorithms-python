```python
#Exercise 4

class HashTable:

    def __init__(self):
        self.MAX = 10
        self.arr = [None for i in range(self.MAX)]
    
    def get_hash(self, key):
        sum = 0
        for char in key:
            sum += ord(char)
        return sum % self.MAX
    
    def __getitem__(self, key):
        index = self.get_hash(key)
        curr_tuple = self.arr[index]

        if curr_tuple == None:
            return "Not Found"
        elif curr_tuple[0] == key:
            return curr_tuple[1]
        else:
            for i in range(index+1, index+self.MAX):
                curr_tuple = self.arr[i % self.MAX]
                if curr_tuple[0] == key:
                    return curr_tuple[1]

    def __setitem__(self, key, value):
        index = self.get_hash(key)
        pair = (key, value)

        if self.arr[index] == None:
            self.arr[index] = pair
        else:
            for i in range(index+1, index+self.MAX):
                new_index = i % self.MAX
                if self.arr[new_index] == None:
                    self.arr[new_index] = pair
                    return

    def __delitem__(self, key):
        index = self.get_hash(key)
        curr_tuple = self.arr[index]

        if curr_tuple[0] == key:
            self.arr[index] = None
        else:
            for i in range(index+1, index+self.MAX):
                curr_tuple = self.arr[i % self.MAX]
                if curr_tuple[0] == key:
                    self.arr[i % self.MAX] = None
```

```python
t = HashTable()
t["march 6"] = 310
t["march 7"] = 420
t["march 8"] = 67
t["march 17"] = 63457

t.arr
```
