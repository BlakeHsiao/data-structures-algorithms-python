```python
#Problem 1

arr = []

with open("nyc_weather.csv", "r") as f:
    for line in f:
        tokens = line.split(',')
        try:
            arr.append(int(tokens[1]))
        except:
            print("Invalid temperature!")

print(sum(arr[0:7])/7)
print(max(arr[0:10]))
```

```python
#Problem 2

dic = {}

with open("nyc_weather.csv", "r") as f:
    for line in f:
        tokens = line.split(',')
        try:
            dic[tokens[0]] = int(tokens[1])
        except:
            print("Invalid temperature!")

print(dic)
print(dic['Jan 9'])
print(dic["Jan 4"])
```

```python
#Problem 3

poem_dic = {}

with open("poem.txt", "r") as f:
    for line in f:
        tokens = line.split(' ')
    
        for word in tokens:
            poem_dic[word] = poem_dic[word] + tokens.count(word)
        

poem_dic
```
